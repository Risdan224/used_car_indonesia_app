```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from webdriver_manager.chrome import ChromeDriverManager

def collecting_data():
    # Collecting data
    vehicle_id = []
    vehicle_title = []
    vehicle_make = []
    vehicle_model = []
    vehicle_body_type = []
    vehicle_year = []
    vehicle_mileage = []
    vehicle_transmission = []
    vehicle_variant = []
    vehicle_type = []
    vehicle_price = []
    vehicle_location = []
    vehicle_link = []

    container = driver.find_elements(By.XPATH, "/html/body/main/div[2]/div/div/div/div/section/article")
    for i in range(len(container)):
        vehicle_id.append(container[i].get_attribute("id"))
        vehicle_title.append(container[i].get_attribute("data-title"))
        vehicle_make.append(container[i].get_attribute("data-make"))
        vehicle_model.append(container[i].get_attribute("data-model"))
        vehicle_body_type.append(container[i].get_attribute("data-body-type"))
        vehicle_year.append(container[i].get_attribute("data-year"))
        vehicle_mileage.append(container[i].get_attribute("data-mileage"))
        vehicle_transmission.append(container[i].get_attribute("data-transmission"))
        vehicle_variant.append(container[i].get_attribute("data-variant"))
        vehicle_type.append(container[i].get_attribute("data-vehicle-type"))
        vehicle_price.append(container[i].text.split('\n')[5])
        vehicle_location.append(container[i].text.split('\n')[-4])
        vehicle_link.append(container[i].get_attribute("data-url"))
    
    
    temp = pd.DataFrame({
        'vehicle_id': vehicle_id,
        'vehicle_title': vehicle_title,
        'vehicle_make': vehicle_make,
        'vehicle_model': vehicle_model,
        'vehicle_body_type': vehicle_body_type,
        'vehicle_year': vehicle_year,
        'vehicle_mileage': vehicle_mileage,
        'vehicle_transmission': vehicle_transmission,
        'vehicle_variant': vehicle_variant,
        'vehicle_type': vehicle_type,
        'vehicle_price': vehicle_price,
        'vehicle_location': vehicle_location,
        'vehicle_location': vehicle_link    
    })
    
    return temp

web = "https://www.carmudi.co.id/mobil-bekas-dijual/indonesia"
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get(web)
driver.maximize_window()
time.sleep(5)

# Open the webiste
# click on last page
driver.find_element(By.XPATH, '//li[@class="last"]').click()
time.sleep(10)
# sorting by "update terakhir"
driver.find_element(By.XPATH, '//div[@class="flexbox__item  guide__input"]').click()
time.sleep(3)
driver.find_element(By.XPATH, '//div[@class="selectize-dropdown-content"]/div[2]').click()
time.sleep(3)

# dropdown_sorting = driver.find_element(By.XPATH, '//*[@id="classified-listings-result"]/div[4]/div/nav/div/div/div/select')
# dropdown_sorting_sel = Select(dropdown_sorting)
# dropdown_sorting_sel.select_by_visible_text("Update Terakhir")
df = collecting_data()
time.sleep(4)

for i in range(551):
    try:
        driver.find_element(By.XPATH, '//li[@class="prev"]').click()
        time.sleep(5)
        df2 = collecting_data()
        df = pd.concat([df, df2], axis=0).reset_index(drop=True)
        time.sleep(1)
    except:
        pass

driver.quit()
```
